```python
# Install necessary packages
!pip install diffusers transformers accelerate torch
!pip install matplotlib

# Import required libraries
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
import os

# Setup device
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Load the Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"  # or "stabilityai/stable-diffusion-2" if you have access
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device == "cuda" else torch.float32)
pipe = pipe.to(device)

# Function to split story into sentences for image generation
def sentence_splitter(script):
    sentences = script.split('.')
    return [sentence.strip() for sentence in sentences if sentence.strip()]

# Function to generate an image based on a prompt
def generate_image(prompt, image_path):
    try:
        image = pipe(prompt).images[0]
        image.save(image_path)
        print(f"Image saved at {image_path}")
    except Exception as e:
        print(f"Failed to generate image for prompt '{prompt}': {e}")

# Directory to save images
save_dir = 'Generated_Images'
os.makedirs(save_dir, exist_ok=True)

# Example story input
script = """
Once upon a time, in a kingdom far away, there was a beautiful princess named Snow White. Her stepmother, the Queen, was envious of her beauty and commanded a huntsman to take Snow White into the forest and leave her there. Instead, the huntsman, moved by Snow White's kindness, let her go.

Lost in the woods, Snow White discovered a small cottage where seven dwarfs lived. They welcomed her warmly and offered her shelter. In return, Snow White took care of their home.
"""  # Replace this with any story text

# Split the script into sentences
sentences = sentence_splitter(script)

# Generate and save images for each sentence
for i, sentence in enumerate(sentences):
    prompt = f"{sentence} in a highly detailed and vivid fantasy style, high-resolution, cinematic lighting."
    image_path = os.path.join(save_dir, f"image_{i+1}.png")
    print(f"Generating image for sentence {i + 1}: {sentence}")
    generate_image(prompt, image_path)

print("Image generation complete. Check the Generated_Images folder for results.")
```
